```python
#gradient decent
for i in range(0,10000):
    yprd = b0 + b1*x1 + b2*x2
    # yprd
    pred = 1 / (1+ np.exp(-yprd))
    # np.round(pred)
    D_b0 = -2 * sum((y_train - pred) * pred * (1 - pred))  
    D_b1 = -2 * sum(x1 * (y_train - pred) * pred * (1 - pred))  
    D_b2 = -2 * sum(x2 * (y_train - pred) * pred * (1 - pred))  
    # Update b0 and b1
    b0 = b0 - 0.01 * D_b0
    b1 = b1 - 0.01 * D_b1
```
